```python
df=pd.read_csv('/content/RELI.csv')
df=df.set_index(pd.DatetimeIndex(df['Date'].values))
df

```

```python
def SMA(data,period=10,column='Close'):
  return data[column].rolling(window=period).mean()

df['SMA10']=SMA(df)
df


```

```python
#moving average
def staterdy(df):

  buy=[]
  sell=[]
  flag=0
  buyprice=0
  #flag 0 means it is okay to buy
  for i in range(0,len(df)):
    if df['SMA10'][i]>df['Close'][i] and flag ==0:
      buy.append(df['Close'][i])
      sell.append(np.nan)
      buyprice=df['Close'][i]
      flag=1

    elif df['SMA10'][i]<df['Close'][i] and flag ==1 and buyprice<df['Close'][i]:
      sell.append(df['Close'][i])
      buy.append(np.nan)
      #we are selling so make buyprice 0 and change flag
      buyprice=0
      flag=0

    else:
      sell.append(np.nan)
      buy.append(np.nan)

  return (buy,sell)
```

```python
strat=staterdy(df)


df['Buy']=strat[0]
df['Sell']=strat[1]

# (np.isnan(df['Buy'][0]))
```

```python
#pmo
p=[]


def pmo(data,p):
  val=0
  for i in range(9,len(data)):
    val=data['Close'][i]-data['Close'][i-9]
    p.append(val)

pmo(df,p)

index_arr=[]
for i in range(9,len(df)):
  index_arr.append(i)

column_val=['Close']
# print(p)
data_arr=np.array(p) 

df1=pd.DataFrame(
    data=data_arr,
    index=index_arr,
    columns=column_val)
df['PMO']=df1

def statergypmo(df):
  flag=0
  buyprice=0  
  pred_buy=[]
  pred_sell=[]
  buyprice=0
  for i in range(0,len(df)):

    if i>=9:
      t=i-9
      s=df['Close'][i]-df['Close'][t]
      if(s>=0) and flag==0:
        # print("predict increase")
        pred_buy.append(df['Close'][i])
        pred_sell.append(np.nan)
        buyprice=df['Close'][i]
        flag=1

      elif(s<0) and flag==1 and  buyprice<df['Close'][i]:
        # print("decrease") 
        pred_buy.append(np.nan)
        pred_sell.append(df['Close'][i])
        buyprice=0
        flag=0
      
      else:
         pred_buy.append(np.nan)
         pred_sell.append(np.nan)

    else:
      pred_buy.append(np.nan)
      pred_sell.append(np.nan)
  return (pred_buy,pred_sell)


# plt.plot()
# plt.plot(df1)
df['Buy_stock']= statergypmo(df)[0]
df['Sell_stock']=statergypmo(df)[1]

# len(df['Buy_stock'])
```

```python
#stockastic

def stockastic_stat(data,p):
  buyprice=0
  buy_stock=[]
  sell_stock=[]
  j=0
  flag=0

  
  for i in range (0,len(data)):
    if i>=4:
        tc=data['Close'][i]
        ln=min(data['Close'][j:i+1])
        hn=max(data['Close'][j:i+1])
       
        j=j+1
        k=(tc-ln)*100/(hn-ln)
        p.append(k)
      

        if k>40:

          buy_stock.append(data['Close'][i])
          sell_stock.append(np.nan)
          buyprice=data['Close'][i]
          flag=1



        elif k<=40:
          
          buy_stock.append(np.nan)
          sell_stock.append(data['Close'][i])
          buyprice=0
          flag=0

        else:
            buy_stock.append(np.nan)
            sell_stock.append(np.nan)


       

    else:
      buy_stock.append(np.nan)
      sell_stock.append(np.nan)

  return (buy_stock,sell_stock,p)    

a=[]
strat_k=stockastic_stat(df,a)
df['buy_stock']=strat_k[0]
df['sell_stock']=strat_k[1]      

# print(df['buy_stock'][25])
```

```python
aco_buy_stock=[]
aco_sell_stock=[]
for i in range(0,10):
  aco_sell_stock.append(np.nan)
  aco_buy_stock.append(np.nan)


def moving_pred(a,buy,sell):
  buy=sell=0

  rem=a-10
  
  for i in range(rem,a):
 
    if not(np.isnan(df['Buy'][i])) and not(np.isnan(df['Buy_stock'][i])) or not(np.isnan(df['buy_stock'][i])):
      buy=buy+1

    if not(np.isnan(df['Sell'][i])) and not(np.isnan(df['Sell_stock'][i])) or not(np.isnan(df['sell_stock'][i])):
      sell=sell+1
  
  

  if buy>sell:
    aco_buy_stock.append(df['Close'][a])
    aco_sell_stock.append(np.nan)
    # print("buy")
    


  elif sell>buy: 
    aco_sell_stock.append(df['Close'][a])
    aco_buy_stock.append(np.nan)
    # print("sell")

  else:
      aco_buy_stock.append(np.nan)
      aco_sell_stock.append(np.nan)


  buy=sell=0

  return (aco_buy_stock,aco_sell_stock)
```
